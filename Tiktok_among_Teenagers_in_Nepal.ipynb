```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('survey_responses (2).csv')
```

```python


# Separate the columns into numerical and string columns
num_cols = ['age_range']
str_cols = [ 'student_or_working', 'tiktok_time', 'tiktok_productivity', 'tiktok_purchase',
            'tiktok_sponsored', 'tiktok_content', 'tiktok_influence', 'tiktok_type', 
            'tiktok_created', 'tiktok_media']

# Convert the string columns to numerical
for col in str_cols:
    df[col] = pd.factorize(df[col])[0]

# Print the first few rows of the dataframe to check the conversion
print(df.head())
```

```python
import matplotlib.pyplot as plt
```

```python
for col in df.columns:
    if col == 'id': # Skip the id column
        continue
    counts = df[col].value_counts()
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
    plt.title(col)
    plt.axis('equal')
    plt.show()
```

```python
# Create a correlation matrix for the dataset
corr_matrix = df.corr()
plt.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.show()
```

```python
# Create a histogram of the age range using seaborn
sns.histplot(df['age_range'], bins=5)

# Create a countplot of the type of tiktok content using seaborn
sns.countplot(data=df, x='tiktok_content')

# Create a scatterplot of tiktok time vs tiktok productivity using seaborn
sns.scatterplot(data=df, x='tiktok_time', y='tiktok_productivity')

# Show all the plots
plt.show()
```
